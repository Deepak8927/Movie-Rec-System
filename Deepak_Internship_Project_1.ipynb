# **Problem Statement: Movie Recommendation System üé¨**

With thousands of movies released across various genres, users often struggle to find a movie that suits their preferences. The goal of this project is to create a Movie Recommendation System that suggests movies based on a user's preferred genre and desired number of recommendations. The system should store a collection of movies categorized by genres and use basic Python data structures (lists and dictionaries) to provide recommendations. Additionally, users should be able to customize their recommendations, add new movies, and search for specific movies.

üìå Problem Statement:
Design a Python-based Movie Recommendation System that:

* ‚úÖ Allows users to select a movie genre (Action, Comedy, Drama, Sci-Fi, Horror, etc.).
* ‚úÖ Lets users decide how many recommendations they want.
* ‚úÖ Provides random movie recommendations from the selected genre.
* ‚úÖ Enables users to add new movies to the stored movie list.
* ‚úÖ Stores and retrieves the movie collection using dictionaries.
* ‚úÖ Allows users to search for a movie by title.

üìå Project Objectives:
* üîπ Genre-Based Movie Recommendations ‚Äì Suggest movies based on user-selected genre.
* üîπ Customizable Number of Recommendations ‚Äì Users can choose how many movie suggestions they receive.
* üîπ Randomized Suggestions ‚Äì Ensure variety by picking random movies from the stored collection.
* üîπ Add New Movies ‚Äì Users can expand the movie list by adding their own favorites.
* üîπ Dictionary-Based Storage ‚Äì Store and update the movie collection in a Python dictionary.
* üîπ Search Functionality ‚Äì Allow users to find a movie by title.

üìå Example User Flow:
* Step 1: User runs the program.
* Step 2: System asks the user to select a genre (e.g., "Comedy").
* Step 3: System asks how many movie recommendations they want (e.g., "5").
* Step 4: System displays 5 random movies from the Comedy genre.
* Step 5: User chooses to add a new movie (e.g., "Superbad" under Comedy).
* Step 6: System updates the movie list and saves it.
* Step 7: User searches for a movie by title to check if it exists in the stored collection.


User Runs the Program

The program starts with a menu that provides different options:
* ‚úÖ Get Movie Recommendations
* ‚úÖ Add a New Movie
* ‚úÖ Search for a Movie
* ‚úÖ Exit the Program

```python
# Storing all the genres as keys and movies as values in a dictionary:
movies = {"Action":["HIT","Pushpa","Maharishi","Yashoda","Sky Force","Varisu","Ram Setu","Vaarasudu","Jailer","Bigil","V","Mahaan","K.G.F","Security","Yodha"],
          "Comedy":["VIP-2","Varisu","KING","Pagalpanti","Aavesham","Laila","Dhamaal","Hum","Shaan","Rush Hour","Vidyapati","Maari","Megamind","Home","Housefull"],
          "Drama":["HIT","Udaan","Shershah","G20","Narappa","Tiger 3","Yodha","Pushpa","Vettaiyan","Sultan","MARCO","Raayan","Antony","Gladiator","NGK"],
          "Sci-fi":["Yashoda","Kalki","The Witch","Inception","Meg 2","Solis","Automata","Rampage","Survive","Godzilla","Divergent","Criminal","2.0","65","Prince"],
          "Horror":["P2","Meg 2","Yosi","Overload","Doom","Burial","Sming","Hellboy","Mandrake","Mandy","Qodrat","Azrael","Crawl","Mojin","211"],
          "Thriller":["Chhori","Babygirl","The lake","Crazxy","The Boy","Shaitan","Conclave","Missing","Pathaan","Murmur","Razzi","Kanguva","KA","Evaru","Forest"],
          "Animation":["Dragon Ball","Naruto","Bleach","Integral","One Piece","Death Note","Turbulence","Gintama","Lookism","Mashle","HLB","Robotech","Tonikkaku","Black Clover","Jujustu Kaisen"],
          "Romance":["MCA","My Fault","Marjaavaan","Magadheera","Laila","Dev","Andaaz","Always","Duplicate","Ishaqzaade","Talaash","Lakshya","Saagar","Nasheeb","Bawaal"]
          }

# Storing all the genres in a list for later use:
genres = list(movies.keys())

# Defining a function to get the recommended movies:
def movie_recommendation():
  import random

  # Showing all the available genres:
  print("Available Genres:")
  for gen in genres:
    print("-",gen)

  genre = input("Enter your preferred genre from the above list: ").title().strip()
  if genre in movies.keys():
    num = int(input(f"How many {genre} movies would you like to see? (Choose between 1-15): "))
    if num in range(1,16):
      recom_movies = random.sample(movies[genre], num)
      print("\nüé¨Recommended Movies:üé¨")
      for i in recom_movies:
        print("-",i)
    else:
      print("‚ùå Invalid input! Please enter a number between 1 to 15. ")
  else:
    print("Sorry, this genre is not available! ")

# Defining a function to add movie in existing genre or new genre:
def add_movie():

  # Showing all the available genres:
  print("Available Genres:")
  for gen in genres:
    print("-",gen)

  genre = input("Enter the genre of the movie from the above list (or add a new genre): ").title().strip()
  movie = input("Enter the movie name: ").title()
  if genre in genres:
    t_movies = [item.title() for item in movies[genre]] #Storing the movies of the asked genre in titlecased format for ease in checking:
    if movie not in t_movies:
      movies[genre].append(movie)
      print(f"‚úÖ '{movie}' added to the '{genre}' genre.")
    else:
      print(f"'{movie}' is already exists in the '{genre}' genre, so it cannot be added.")
  else:
    movies[genre] = [movie]
    print(f"‚úÖ '{movie}' is added to the {genre} genre.")

# Defining a function to find the searched movie:
def search_movie():
  movie = input("Enter the movie title to search: ").title()
  titlecase_movies = {k: [item.title() for item in v] for k, v in movies.items()} #Storing all the movies in title-cased format for ease in checking:
  genre = [key for key, value_list in titlecase_movies.items() if movie in value_list] #Getting the genre from movies dictionary of the searched movie in a list:
  if genre != []:
    print(f"'{movie}' is present in '{genre[0]}' genre.")
  else:
    print(f"'{movie}' is not found.")

# Using a while loop to run this code infinite no. of times:
while True:
  print("""
  üé¨Movie Recommendation Systemüé¨
  1. Get Movie Recommendations
  2. Add a New Movie
  3. Search for a Movie
  4. Exit
  """)

  choice = int(input("Enter your choice: "))

  if choice == 1:
    movie_recommendation()

  elif choice == 2:
    add_movie()

  elif choice == 3:
    search_movie()

  elif choice == 4:
    break

  else:
    print("‚ùå Invalid input! Please enter the valid input.")
```
